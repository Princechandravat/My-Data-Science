# Data Cleaning-

```python
#Identify the NaN values-

electric_vehicles.isnull().sum()
```

```python
# handle missing values-

electric_vehicles = electric_vehicles.dropna(subset=['fast_charge_port','cargo_volume_l','model'])

electric_vehicles['number_of_cells']=electric_vehicles['number_of_cells'].fillna(electric_vehicles['number_of_cells'].mean()) # fill with mean

electric_vehicles['torque_nm']=electric_vehicles['torque_nm'].fillna(electric_vehicles['torque_nm'].median()) #fill with median

electric_vehicles['fast_charging_power_kw_dc']=electric_vehicles['fast_charging_power_kw_dc'].fillna(electric_vehicles['fast_charging_power_kw_dc'].median())  #fill with median 

electric_vehicles['towing_capacity_kg']=electric_vehicles['towing_capacity_kg'].fillna(electric_vehicles['towing_capacity_kg'].mean())  #fill with mean
```

```python
# convert 'Cargo_volume_l' to int- 

electric_vehicles['cargo_volume_l'] = pd.to_numeric(
    electric_vehicles['cargo_volume_l'].str.extract(r'(\d+)', expand=False),
    errors='coerce'
).fillna(0).astype(int)
```

```python
#finding null values-
electric_vehicles.isnull().sum()
```

```python
#Remove Duplicated if have anny-

electric_vehicles.drop_duplicates().head()
```

```python
# convert to lowercase & remove whitespace for easy summarizing-

cols = ['segment', 'car_body_type']
for col in cols:
    electric_vehicles[col] = electric_vehicles[col].str.strip().str.lower()  # removing whitespace or convert to lowercase
```

```python
electric_vehicles[['segment','car_body_type']]
```

```python
# Replace 0 with NaN value in 'towing_capacity_kg' -

electric_vehicles['towing_capacity_kg'].replace(0,np.nan)
```

```python
# Fastest Car by top speed-

fastest_by_speed=electric_vehicles[['brand','model','top_speed_kmh']].sort_values(by='top_speed_kmh',ascending = False).head()
fastest_by_speed
```

```python
#Plotting horizontally bar chart-

fastest_by_speed['label'] = fastest_by_speed['brand'] + ' ' + fastest_by_speed['model']

# Plot
plt.figure(figsize=(7, 4))
plt.barh(fastest_by_speed['label'],fastest_by_speed['top_speed_kmh'],color='#1D2549')
plt.xlabel('Speed in KM/H')
plt.title('Top 5 Fastest Electric Cars by top speed')
plt.gca().invert_yaxis()  # Fastest at top
plt.tight_layout()
plt.show()
```

```python
#Fastest car by 0-100 kmh by seconds-

fastest_by_acceleration = electric_vehicles[['brand','model','acceleration_0_100_s']].sort_values(by='acceleration_0_100_s',ascending=True).head()
fastest_by_acceleration
```

```python
fastest_by_acceleration['label']=fastest_by_acceleration['brand'] + ' ' + fastest_by_acceleration['model']


plt.figure(figsize=(7,4))
plt.barh(fastest_by_acceleration['label'],fastest_by_acceleration['acceleration_0_100_s'],color='wheat')
plt.xlabel('0â€“100 km/h Acceleration Time (seconds)')
plt.title('Top 5 Fastest Electric Cars by 0â€“100 km/h Acceleration Time"')
plt.gca().invert_yaxis()  # Fastest at top
plt.tight_layout()
plt.show()
```

```python
# Distribution of battery_capacity_kWh by brands-

#figure-
plt.figure(figsize=(20,10))
x=electric_vehicles[['brand','model','battery_capacity_kWh']]
plt.bar(x['brand'],x['battery_capacity_kWh'],color='seagreen',width=0.8,label='Center Aligned')
plt.title('Distribution of battery_Capacity_kWh',fontsize=30)
plt.ylabel('Battery Capacity (kWh)',fontsize=25)
plt.xlabel('Car Brands',fontsize=25)
plt.xticks(rotation=70)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Top speed Cars-

top_speeds=electric_vehicles['top_speed_kmh']
plt.figure(figsize=(10, 6))
plt.hist(top_speeds, bins=20, color='orange', edgecolor='black')
plt.title("Distribution of Top Speed")
plt.xlabel("Top Speed (km/h)")
plt.ylabel("Number of Vehicles")

# Add vertical line at mean
mean_speed = top_speeds.mean()
plt.axvline(mean_speed, color='red', linestyle='--', linewidth=2, label=f"Mean: {mean_speed:.1f} km/h")

# Add legend
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
