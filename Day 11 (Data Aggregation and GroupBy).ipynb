```python
import numpy as np
import pandas as pd

#Reading a CSV file-

movie=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\imdb-top-1000.csv")
movie.head()
```

```python
genre=movie.groupby('Genre')
genre
```

```python
#Sum of Genre - using Groupby function().

genre.sum().head()
```

```python
#find the top 3 geners by total earning-

genre.sum()['Gross'].sort_values(ascending=False).head(3) #top 3 highest earning geners.
# first sum of all generes with Gross then sort with descending ording .
```

```python
#finding the genre with highest avg imdb rating-

movie.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
#What is the highest IMDb rating for each year-

movie.groupby('Released_Year')['IMDB_Rating'].max()
# Groupby 'Released_Year' and max() on IMDB_Rating.
```

```python
# How many movies have a Meta score above the average Meta score-

#movies with the avg Metascore-
avg_metascore=movie['Metascore'].mean()
avg_metascore

#Filter movies with the high score than average_metascore.
above_avg=movie[movie['Metascore']>avg_metascore]
above_avg

#count the movies high metascore than average metascore.
count_above_avg_metascore=above_avg.shape[0]

print("Average Meta Score:- ",avg_metascore)
print("No. of Movie with above metascore than average metascore:- ",count_above_avg_metascore)

above_avg
```

```python
# Among the top 5 genres with the most movies, which one has the highest average rating-

#Top 5 genres by IMDB_Rating-
top_5_genres = movie['Genre'].value_counts().head(5).index.tolist()
top_5_genres

filter=movie[movie['Genre'].isin(top_5_genres)]
avg=filter.groupby('Genre')['IMDB_Rating'].mean()
avg
```

```python
# Using .agg()-  GroupBy columns with multiple aggregations-

# For each director, find the average IMDb rating and total runtime of their movies.

movie.groupby('Director').agg({
    'IMDB_Rating':'mean',
    'Runtime':'sum'
})
```

```python
# Q)- For each actor listed in the Star1 column, calculate:
#(a)- The average IMDb rating of all their movies
#(b)- The total number of movies they appeared in as lead
#(c)- The minimum Meta score among their movies


movie.groupby('Star1').agg({
    'IMDB_Rating':'mean',
    'Series_Title':'count',
    'Metascore':'min'
})
```

```python
movie.groupby('Star1').agg({
    'IMDB_Rating':['mean','max','min'],  #find min/max/mean of IMDB_Rating.
    'Series_Title':['count'],            #Finding count of Series_Title
    'Metascore':['min','mean','max']      #Min/max/mean of Metascore
})
```

```python
# Using .agg()-  Using list of function for one column-

movie.groupby('Genre')['Runtime'].agg(['min','max','mean','sum'])
```

```python
#apply()-

movie.groupby('Genre').apply(min)  #applying min()
```

```python
#Create a norm_rating function for calculating normalization


#Creating a function for print Normalization-
def normal(group):
    group['norm_rating']=group['IMDB_Rating']-group['IMDB_Rating'].min()/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group

normalization=movie.groupby('Genre').apply(normal)    #creating a variable for apply function.

normalization  
```
