```python
#loading titanic dataset-
titanic=sns.load_dataset('titanic')
titanic
```

```python
# Distribution of fare by class using stripplot-

sns.stripplot(data=titanic,x='class',y='fare',jitter=0.3,hue='sex')
plt.title('Distrbution of Fare by class')
```

```python
# also create with figure level function-

sns.catplot(data=titanic,x='class',y='fare',kind='strip',jitter=0.3,hue='sex')
```

```python
#Swarm plot-
```

```python
sns.swarmplot(x='pclass', y='age', data=titanic,size=3.5,hue='sex')
```

```python
#Using figure level function-

sns.catplot(data=titanic,x='pclass',y='age',hue='sex',kind='swarm',size=3.5)
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
#PairGrid-

x=sns.PairGrid(iris,hue='species')
x.map(sns.scatterplot)
```

```python
#Pairplot-

y=sns.PairGrid(iris,hue='species')
y.map_diag(sns.histplot)
y.map_offdiag(sns.scatterplot)
```

```python
#PairGrid with customization -

m=sns.PairGrid(data=iris,hue='species')
m.map_diag(sns.histplot)
m.map_upper(sns.kdeplot)
m.map_lower(sns.scatterplot)
```

```python
# vars in PairGrid-

m=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])  # vars used for specify columns
m.map_diag(sns.histplot)
m.map_upper(sns.kdeplot)
m.map_lower(sns.scatterplot)
```

```python
g=sns.JointGrid(tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.boxplot)
```

```python
g=sns.JointGrid(tips,x='total_bill',y='tip',hue='sex')
g.plot(sns.scatterplot,sns.barplot)
```

```python
g=sns.JointGrid(tips,x='total_bill',y='tip',hue='sex')
g.plot(sns.kdeplot,sns.violinplot)
```
