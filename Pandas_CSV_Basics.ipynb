# Writing CSV File in Pandas:-

```python
# Creating CSV DataFrame using stringIO-

import pandas as pd
from io import StringIO

# Define the CSV data as a multiline string
data = """Name,Age,City
Prince,20,Delhi
Keshav,21,Gurugram
Hritik,22,Jaipur
Devesh,20,Mumbai
Kartik,22,Agra"""

# Create an in-memory file-like buffer
csv_buffer = StringIO(data)

# Read the CSV data into a pandas DataFrame
df = pd.read_csv(csv_buffer)

# Display the resulting DataFrame
df
```

```python
#Basic CSV Writing-
import pandas as pd

#Creating dataFrame
data = {
    'Name': ['Prince chandravat', 'Shruti khapra', 'Vibhor sharma'],
    'Age': [25, 30, 22],
    'City': ['Delhi', 'Mumbai', 'Bangalore']
}
df = pd.DataFrame(data)

#Write to csv
df.to_csv('basic_data.csv')

#Check
df.head()
```

```python
#Writing without indexing-

data={
    'Name':['Prince','Gaurav','Sonakshi'],
    'Age':[20,22,23],
    'city':['Delhi','Mumbai','Ahemdabad']}

df1=pd.DataFrame(data)

df1.to_csv("without_indexing.csv",index=False)
print(df1.head())
```

```python
# Writing again with index for demo
df.to_csv('custom_index.csv')

# Reading with index_col
df_indexed = pd.read_csv('custom_index.csv', index_col=0)
print(df_indexed)
```

```python
# Read only Name and City columns
df_cols = pd.read_csv('without_index.csv', usecols=['Name', 'City'])
print(df_cols)
```

```python
# Write with a pipe separator
df.to_csv('pipe_sep.csv', sep='|', index=False)

# Read with pipe separator
df_pipe = pd.read_csv('pipe_sep.csv', sep='|')
print(df_pipe)
```
