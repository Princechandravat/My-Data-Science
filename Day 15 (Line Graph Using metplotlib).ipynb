```python
#from a dataframe-

batsman=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\sharma-kohli.csv")
batsman
plt.plot(batsman['index'],batsman['V Kohli'])  #using plot function and plot index & v kohli score.
```

```python
# Multiple plotting-
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Virat Kohli Vs Rohit Sharma Comparison')  #title providing.
```

```python
#Xlabel-Ylabel-

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs scored')   #ylabel use for Provide name on Y-Axis
```

```python
#Change the color of lines  using (Hex) or by color name-

plt.plot(batsman['index'],batsman['V Kohli'],color='#738CA0')   #using color function
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#3D9F65')   #using color function

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
```

```python
#linestyle-

plt.plot(batsman['index'],batsman['V Kohli'],color='#738CA0',linestyle='dotted')   
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#3D9F65',linestyle='dotted')   

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#103AF1',linestyle='dashdot')   #dashdot linestyle
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#FAC209',linestyle='dashdot') #dashdot linestyle

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
```

```python
#adjust width of line using (linewidth)-

plt.plot(batsman['index'],batsman['V Kohli'],color='#FA0909',linewidth=5)   
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#103AF1',linewidth=6)

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
plt.grid()  #for grid
```

```python
#Providing label to each line-

plt.plot(batsman['index'],batsman['V Kohli'],color='#FA0909',linewidth=5,label='Virat kohli')   
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#103AF1',linewidth=6,label='Rohit Sharma')

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
plt.legend()
plt.grid()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#FA0909',linewidth=5,label='Virat kohli')   
plt.plot(batsman['index'],batsman['RG Sharma'],color= '#103AF1',linewidth=6,label='Rohit Sharma')

plt.title('Virat Kohli Vs Rohit Sharma Comparison')
plt.xlabel('Season')  #xlabel use for Provide name on X-Axis
plt.ylabel('Runs Scored')   #ylabel use for Provide name on Y-Axis
plt.legend(loc='best')  #label location can be change to a specific place. like- upper right,upper left,left,right,centre etc.
plt.grid()
```

```python
#To see the line between a specific range-  

# Limiting axes-


# for y-axis use ylim()-
#for x-axis use xlim()- 

Year=[2019,2020,2021,2022,2023,2024]
Price=[45000,51500,52100,44300,52100,5000000]

plt.plot(Year,Price)
print(plt.ylim(0,76000))  #for y-axis
plt.grid()
```

```python
Year=[2019,2020,2021,2022,2023,2024]
Price=[45000,51500,52100,44300,52100,5000000]

plt.plot(Year,Price)
print(plt.ylabel('Price'))
print(plt.xlabel('Year'))
print(plt.ylim(30000,70000))  #ylim() for y-axis
print(plt.xlim(2019,2022))  #xlim() for x-axis
plt.grid()
```
