# 1- Data Handling(Missing Values) using dropna & fillna:-

```python
import pandas as pd
```

```python
#Reading a CSV file-

#Titanic passengers csv.
csv=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\titanic.csv")  
csv
```

```python
csv.describe()
```

```python
#finding missing values in boolean:-

csv.isnull()        #True/False
```

```python
#finding missing values by columns:-

csv.isnull().sum() 
```

```python
#Drop Row with missing values-

dropped=csv.dropna()  #Drop the (NaN) Rows.
dropped
```

```python
#Fill Missing Vaues (using fillna):-

# 1-> Replacing NaN values with 0

filled=csv.fillna(0)   #Replace all NaN values with 0
filled
```

```python
# Fill with different values in column

custom_csv = csv.fillna({
    'Age': csv['Age'].mean(),     #Replace all the NaN values in Age( with average of age).
    'Cabin': 'Unknown',           #Replace all the NaN values in Cabin with 'Unknown'.
    'Fare': csv['Fare'].median()  #Replace all the NaN values in Fare( with median of Fare).
})
custom_csv
```

```python
# Forward fill-

forward_filled=csv.fillna(method='ffill')  #It replace the NaN values with forward value
forward_filled
```

```python
# Backward fill-

backward_filled=csv.fillna(method='bfill')  #It replace the NaN values with backward value

backward_filled
```

```python
#Now save this data to a new csv with cleaned data-

custom_csv.to_csv("titanic_cleaned.csv",index=False)     #Cretating a new csv file Without Index.
```

```python
#reading a movie csv file for filtering and operating boolean functions:-

movies=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\movies.csv")
movies.sample(1)
```

```python
# Show all movies with an IMDb rating greater than 8.0 and votes>10000

movie=movies[(movies['imdb_rating']>8.0) & (movies['imdb_votes']>10000)]    #all the movies above 8.0 imdb rating with 10000 above rating.
print("Total no. of movies with 8.0 above ratings and votes over 10000 is==>", movie.shape[0])
movie
```

```python
#Filter action movies with 8.0 or above imdb rating:-

mask1=movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
mask2=movies['imdb_rating']>= 8.0

result=movies[mask1 & mask2]
result
```

# Using isin() , .str.contains() & isna():-




```python
#.Str.contains()- Using function-

#Filter action movies with 8.0 or above imdb rating:-

mask1=movies['genres'].str.contains('Action')
mask2=movies['imdb_rating']>= 8.0

result=movies[mask1 & mask2]
result
```

```python
# another example of .str.contains()-

#Filter movies akshay kumar & salman khan together-

mask1=movies['actors'].str.contains('Akshay kumar',case=False, na=False)  
mask1=movies['actors'].str.contains('salman khan',case=False, na=False)

#case = False--> it handles the lower,upper and all size capital letters
#na = False--> it false the NaN values

result= movies[mask1 & mask2]
result
```
