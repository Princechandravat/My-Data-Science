```python
import pandas as pd 
```

```python
movies=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\imdb-top-1000.csv")
movies

```

```python
#astype-

movies['Runtime'].astype('int16')  #this function convert the data in differents bits
```

```python
#between()-

movies['Runtime'].between(120,150)  #Runtime between 120-150 movies with boolean values.
```

```python
#Drop_duplicate-

duplicates_drop=movies['Director'].drop_duplicates()  #drops the directors which have more than 1 movie
print(duplicates_drop)
print()
print("The no. after dropped duplicated  directors-->",duplicates_drop.size)  #size after dropped duplicate values.
```

```python
#isnull()-

movies.isnull()  #this function show NaN values in the form of booleans
```

```python
#isnull().sum()-

movies.isnull().sum()  #it show sum of nan values in each columns.
```

```python
#dropna()-

movies.dropna()  #this function drops all the mssing values.
```

```python
#fillna()-   

movies.fillna(0)  #this function fill the Nan values with a specific value like-0

movies.fillna(method='ffill') #this used for fill the Nan values with forward value

movies.fillna(method='bfill') #this function used for fill the NaN value with Backward value
```

```python
#isin()-

#Checks whether each element in a Series is contained in a given list or set, returning a Boolean Series

movies[movies['Released_Year'].isin(['2020','1972'])]
```

```python
#apply-

# Applies a given function to each element in the Series and returns a new Series with the results-
movies['Director'].apply(lambda x: x.split()[0])  #use a function for just first name of director by apply() function.
```

```python

new=movies['Series_Title'][1]='Titanic'
new

#it changes permanent in original data- 'movies' so because we copy() function it creates a copy function and which dont effect on original data-

#Copy-

copy=movies['Series_Title'][1]='The Godfather'
copy
```
