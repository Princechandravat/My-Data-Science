# Notebook 23 (Plot Scatter & Line chartusing Seaborn)-

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Scatter plot-
```

```python
#loading a dataset-
df=sns.load_dataset('tips')
df
```

```python
#Scatter plot (using axes level function)-
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# Scatter plot (using figure level function)-
```

```python
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
employee_data=pd.read_csv(r"C:\Users\Prince Chandravat\OneDrive\Videos\Documents\DATASETS\employee salary dataset\synthetic_salary_dataset.csv")
employee_data
```

```python
#Positions with avg salary-

position=employee_data.groupby('Position')['Salary(INR)'].mean().reset_index()
position
```

```python
sns.relplot(data=employee_data,x=employee_data['Position'],y=employee_data['Salary(INR)'],kind='scatter',size='YearsExperience')
plt.xticks(rotation=85)
plt.show()
```

```python
#Top High salray positions-

employee_data.groupby('Position')['Salary(INR)'].mean().sort_values(ascending=False).reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=employee_data, x='Position', y='Salary(INR)',align='edge',color='teal')
plt.xticks(rotation=85)
plt.title("Salary Distribution by Position")
plt.show()
```

```python
#loading a dataset -
gap=px.data.gapminder()
gap
```

```python
india=gap[gap['country']=='India']
india
```

```python
#ploting a line plot-  (by axes leve function)

sns.lineplot(data=india,x='year',y='lifeExp',hue='country')
plt.grid()
```

```python
# (by figure level function)-

sns.relplot(data=india,x='year',y='lifeExp',kind='line')
plt.grid()
```
