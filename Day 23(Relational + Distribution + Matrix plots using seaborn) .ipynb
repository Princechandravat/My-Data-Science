```python
#Cleaning the dataset-

car['horsepower'] = pd.to_numeric(car['horsepower'],errors='coerce')

origin_map = {1: 'USA', 2: 'Europe', 3: 'Japan'}
car['origin'] = car['origin'].replace(origin_map
                                     )
df.dropna(inplace=True)  # Drop rows with any missing values
```

# Some practice of relational plotting-

```python
# relationship between weight and mpg-


plt.figure(figsize=(14,7))
sns.relplot(kind='scatter',data=car,x='mpg',y='weight',hue='origin',size='weight')

plt.title('Fuel Efficiency (mpg) vs Car Weight')
plt.grid()
plt.show()
```

```python
# relationship between horsepower and mpg-

plt.figure(figsize=(14,6))
sns.relplot(kind='scatter',data=car,x='horsepower',y='mpg',size='weight',hue='origin',style='cylinders')
```

```python
# Clustering of Cars by Origin: Weight vs. Acceleration (Size = MPG)-

#figure-
plt.figure(figsize=(10, 6))

#plotting-
sns.scatterplot(data=car,x='weight',y='acceleration',hue='origin',size='mpg',sizes=(40, 200),palette='deep',alpha=0.7)
plt.title('Clustering of Cars by Origin: Weight vs. Acceleration (Size = MPG)')

plt.grid(True)
plt.show()
```

```python
# Relationship between horsepower and mpg-

plt.figure(figsize=(10, 6))

sns.scatterplot(data=car,x='horsepower',y='mpg',size='acceleration',sizes=(30, 300),hue='origin',marker='d')

plt.grid()
plt.show()
```

# Matrix plots-

1)- Heatmap plot
2)- Cluster plot

```python
#Heatmap-
```

```python
gap.sample(10)
```

```python
x=gap.pivot_table(index='country',columns='year',values='lifeExp')
x
```

```python
#creating a heatmap of country and lifeExp year-

#figure-
plt.figure(figsize=(15,10))
sns.heatmap(x,cmap='ocean')
```

```python
#Only European country heatmap-

europe=gap[gap['continent']=='Europe'].pivot_table(index='country',columns='year',values='lifeExp')
europe
```

```python
#Plotting heatmap-

plt.figure(figsize=(15,12))  
sns.heatmap(data=europe,cmap='YlGnBu',linewidth=0.5,annot=True)
```

```python
#Asian countries-

asian=gap[gap['continent']=='Asia'].pivot_table(index='country',columns='year',values='lifeExp')
asian
```

```python
plt.figure(figsize=(15,12))  
sns.heatmap(asian,linewidth=0.5,annot=True)
plt.show()
```
