```python
# Stock Prices Line Chart
# Plot the stock prices of a company over 7 days.


prices = [250, 255, 253, 260, 262, 268, 270]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.plot(days,prices,marker='o',markersize=10,linewidth=5)
plt.grid()
plt.xlabel('days')
plt.ylabel('Price')
plt.title('Stock Prices Line Chart')
plt.tight_layout()
```

```python
# Q2- Temperature Comparison
# Plot the temperature (in °C) of two cities over 12 months in one line graph with different colors and legends.

months=['January','February','March','April','May','June','July','August','September','October','November','December']

City1_temperature=[5, 8, 14, 20, 25, 30, 33, 32, 28, 20, 12, 6]  #Delhi
City2_temperature=[0, 2, 8, 15, 20, 25, 27, 26, 22, 15, 7, 2]  #Shimla


plt.plot(months,City1_temperature,color='green',linewidth=5,marker='o',markersize=10,label='Delhi')
plt.plot(months,City2_temperature,color='orange',linewidth=5,marker='o',markersize=10,label='Shimla')
plt.grid()
plt.xlabel('months',fontsize=12)
plt.ylabel('Temperature in Celsius',fontsize=12)
plt.title('Temperature Comparison',fontsize=18)
plt.legend()
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Q-3 Covid Cases Trend-
# Plot daily new Covid-19 cases over a month using a line graph. Use xlabel, ylabel, and title.


# Sample Data: Daily cases for 30 days
days = list(range(1, 31))
covid_cases = [100, 120, 130, 110, 150, 170, 200,
               180, 160, 190, 210, 220, 240, 250,
               230, 220, 200, 210, 230, 250, 270,
               290, 300, 280, 260, 240, 230, 220,
               210, 200]

# Plotting the line graph
plt.figure(figsize=(10, 5))
plt.plot(days, covid_cases, color='red', marker='o', linestyle='-', linewidth=2)

# Adding labels and title
plt.title('Daily New COVID-19 Cases Over a Month', fontsize=14)
plt.xlabel('Day of the Month', fontsize=12)
plt.ylabel('New COVID-19 Cases', fontsize=12)
plt.grid(True)
plt.xticks(days)
# Display the plot
plt.show()
```

```python
#Q-4 Sales Data Bar Chart
# Plot the monthly sales (in ₹) for a store:

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [15000, 18000, 22000, 21000, 25000]

plt.bar(months,sales,color='Red')  #plotting a bar raph
plt.xlabel('Months',fontsize=12) #labelling x-axis
plt.ylabel('Sales',fontsize=12)  #labelling y-axis
plt.title('Sales data of store')
```

```python
#Q- 5-Horizontal Bar Chart-
# Display a horizontal bar chart of different smartphone brands and their market share-

smartphone_brands = ['Samsung', 'Apple', 'Motorola', 'Vivo', 'Oppo', 'Xiomi', 'Nokia', 'Realme', 'OnePlus']
market_share = [22,19,5,8,7,10,3,4,5]

plt.barh(smartphone_brands,market_share,color='#080E56')
plt.xlabel('SmartPhone Brands',fontsize=12)
plt.ylabel('Market Share',fontsize=12)
plt.title('smartphone market share',color='r',fontsize=14)
plt.show()
```

```python
# Q-6-Grouped Bar Chart-
# Compare the marks of two students in 5 subjects using a grouped bar chart-

# Subjects
subjects = ['Math', 'Science', 'English', 'History', 'Computer']

# Marks of two students
student1 = [85, 90, 78, 92, 88]
student2 = [80, 85, 82, 89, 90]

x=np.arange(len(subjects))

bar_width=0.35 #width of bar

plt.figure(figsize=(12,5))  #adjust the size of bar

plt.bar(x-bar_width/2,student1,width=bar_width,label='Students1',color='#14140F') 
plt.bar(x+bar_width/2,student2,width=bar_width,label='Students2',color='salmon')

plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.xlabel('Subjects',fontsize=14)  #providing x-label naming
plt.ylabel('Marks',fontsize=14)     #providing y-label naming

plt.xticks(x,subjects) 
plt.title('Comparison of marks between student1 and student2',fontsize=18) #title name 


plt.show()  #show the  bar graph
```

```python
#Q7-Height vs Weight
# Plot a scatter plot to show the relationship between height and weight of 10 individuals.


#figure-
plt.figure(figsize=(10,5))
# Sample data: Height (in cm) and Weight (in kg)
heights = [150, 155, 160, 165, 170, 175, 180, 185, 190, 195]
weights = [45, 50, 52, 58, 63, 67, 72, 76, 80, 85]

plt.scatter(heights,weights,edgecolor='black',s=100)
plt.grid(True,linestyle='--')
plt.show()
```

```python
expense=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\monthly_expenses.csv")
expense
```

```python
# Smartphones and shares-
brands = ['Samsung', 'Apple', 'Xiaomi', 'OnePlus', 'Others']
shares = [28, 25, 20, 15, 12]

#creating a pie chart-
plt.pie(shares,labels=brands,autopct='%1.1f%%')
plt.title('Smartphones market share',fontsize=15)
plt.show()
```
