# Bar chart:-

```python
batsman=pd.read_csv("C:\\Users\\Prince Chandravat\\Downloads\\batsman_season_record.csv")
batsman
```

```python
plt.bar(batsman['batsman'],batsman['2017'])
```

```python
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=0.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.2)
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2017'],width=0.2)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.show()
```

```python
#StackBar Chart-


plt.bar(batsman['batsman'],batsman['2017'],label='2017')
plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2017'],label='2016')
plt.bar(batsman['batsman'],batsman['2015'],bottom=batsman['2016']+batsman['2017'],label='2015')

plt.legend()
plt.show()
```
