```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading ipl dataset-

ipl=pd.read_csv(r"C:\Users\Prince Chandravat\OneDrive\Videos\Documents\DATASETS\IPL Dataset\most_runs_average_strikerate.csv")
ipl.head()
```

```python
# plotting bar graph of top 5 highest run scorer-

#data-
batsman = ipl['batsman'].head(10)
runs = ipl['total_runs'].head(10)

#figure-
plt.figure(figsize=(10,4))

# creating-
plt.bar(batsman,runs,color='brown',width=0.5)
plt.xlabel('Batsman names',fontsize=15)
plt.ylabel('Score',fontsize=15)
plt.title('Top 10 Batsman with most runs in IPL')
plt.xticks(rotation=30)
plt.show()
```

```python
#plotting a Line chart-

matches=pd.read_csv(r"C:\Users\Prince Chandravat\OneDrive\Videos\Documents\DATASETS\IPL Dataset\matches.csv")
matches.head()
```

```python
#Line chart of total matches of ipl each season-

matches_per_year = matches['Season'].value_counts().sort_index()
matches_per_year

#figure-
plt.figure(figsize=(11,5))

plt.plot(matches_per_year.index,matches_per_year.values)
plt.grid()
plt.xticks(rotation=70)
plt.xlabel('Season',fontsize=13,color='r')
plt.ylabel('Matches',fontsize=13,color='r')
plt.title('Total IPL Matches each Season',fontsize=16,color='r')
plt.show()
```

```python
#ceating subplots-
fig, ax=plt.subplots(nrows=2,ncols=1,figsize=(20,18))

#Creating a sample ipl data-
sample=ipl.head(100).sample(10)
sample2=ipl.head(200).sample(50)

#Annotation-
for i in range(sample2.shape[0]):
    ax[0].text(sample2['average'].values[i],sample2['strikerate'].values[i],sample2['batsman'].values[i])

    
#average with strikerate -
ax[0].scatter(sample2['average'],sample2['strikerate'],color='#1573F8',marker='^')

#batsman with strikerate-
ax[1].bar(sample['batsman'],sample['strikerate'],align='center')

plt.setp(ax[1].get_xticklabels(), rotation=45, ha='right')
        
#labelling -
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strikerate')
ax[0].set_title('Strikerate Vs Average') 
ax[0].grid() 

ax[1].set_xlabel('Batsman')
ax[1].set_ylabel('strikerate')
ax[1].set_title('Batsman vs strikerate')

plt.tight_layout()
plt.show()
```

```python
#ceating subplots-
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(20,18))

#Creating a sample ipl data-
sample=ipl.head(100).sample(10)
sample2=ipl.head(200).sample(50)

#Annotation-
for i in range(sample2.shape[0]):
    ax[0,0].text(sample2['average'].values[i],sample2['strikerate'].values[i],sample2['batsman'].values[i])

    

ax[0,0].scatter(sample2['average'],sample2['strikerate'],color='#1573F8',marker='^')
ax[0,1].bar(sample['batsman'],sample['strikerate'],align='center')
ax[1,0].hist(sample2['strikerate'],bins=[100,110,120,130,140,150,160,170,180,190,200,250],color='g')
ax[1,1].scatter(sample2['total_runs'],sample2['strikerate'],color='#1573F8',marker='^')

plt.setp(ax[0,1].get_xticklabels(), rotation=45, ha='right')

#labelling -
ax[0,0].set_xlabel('Average',fontsize=20)                      
ax[0,0].set_ylabel('Strikerate',fontsize=20)
ax[0,0].set_title('Strikerate Vs Average',fontsize=20) 
ax[0,0].grid() 

ax[0,1].set_xlabel('Batsman',fontsize=20)
ax[0,1].set_ylabel('strikerate',fontsize=20)
ax[0,1].set_title('Batsman vs strikerate',fontsize=20)

ax[1,0].set_xlabel('Strikerate',fontsize=20)

ax[1,1].set_xlabel('Total runs',fontsize=20)
ax[1,1].set_ylabel('Strikerate',fontsize=20)
ax[1,1].set_title('Runs Vs Strikerate',fontsize=20)


plt.tight_layout()
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z=np.cos(xx) + np.tan(yy)

#figure-
fig = plt.figure(figsize=(15,5))

ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z)

fig.colorbar(p)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

#figure-
fig = plt.figure(figsize=(15,8))

ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z)
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z=xx**2+yy**2

#figure-
fig = plt.figure(figsize=(15,7))

ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='grey')

fig.colorbar(p)
plt.show()
```
